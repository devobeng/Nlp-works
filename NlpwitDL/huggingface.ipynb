```python
%%time
import pandas as pd
from transformers import pipeline
from transformers import logging
logging.set_verbosity_error()
pd.set_option("display.max_colwidth", None)
```

```python
df=pd.read_excel('../Data/Popchip_Reviews_Sentiment.xlsx').head(30)
```

```python
df.head()
```

```python
sentimental_analyser=pipeline('sentiment-analysis', model="distilbert/distilbert-base-uncased-finetuned-sst-2-english", device=-1, truncation=True)
```

```python
sentimet_scores=df.Text.apply(sentimental_analyser)
```

```python
sentimet_scores
```

```python
df['Label_HF']=sentimet_scores.apply(lambda x: x[0]['label'])
df['Scores_HF']=sentimet_scores.apply(lambda x: x[0]['score'])
df['Sentiment_HF']=df.apply(lambda row: row['Scores_HF'] if row['Label_HF']=='POSITIVE' else -row['Scores_HF'], axis=1)
```
