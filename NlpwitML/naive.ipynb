```python
reviews['Text_Clean']=piplineline.clean_and_normilize(reviews.Text)
```

```python
reviews.head()
```

```python
cv_df=cv.fit_transform(reviews.Text_Clean)
```

```python
df_cv=pd.DataFrame(cv_df.toarray(), columns=cv.get_feature_names_out())
```

```python
df_cv
```

```python
X_train, X_test, y_train, y_test=train_test_split(cv_df, reviews.Priority, test_size=0.2, random_state=42)
model=MultinomialNB()
model.fit(X_train, y_train)
y_pred=model.predict(X_test)
print("Accuracy", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
