```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
df=pd.read_csv('../Data/movie_reviews.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
def get_sentiment(text):
    anlyser=SentimentIntensityAnalyzer()
    return anlyser.polarity_scores(text)['compound']
```

```python
df['vardar_sentiment']= df['movie_info'].apply(get_sentiment)
```

```python
df.head()
```

```python
df[['vardar_sentiment', 'movie_title']].sort_values(by="vardar_sentiment", ascending=False)
```

```python
df[['vardar_sentiment', 'movie_title']].sort_values(by="vardar_sentiment", ascending=False).tail(10)
```
