```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
import piplineline
```

```python
df= pd.read_csv('../Data/movie_reviews.csv')
```

```python
df.head()
```

```python
df['clean_movie_info']= piplineline.clean_and_normilize(df.movie_info)
```

```python
df.head()
```

```python
df.director_gender.value_counts()
```

```python
cv=CountVectorizer(stop_words='english', min_df=.1)
df_cv=cv.fit_transform(df.clean_movie_info)
```

```python
X_train, X_test, y_train, y_test=train_test_split(df_cv,df.director_gender,  test_size=0.2, random_state=42)
```

```python
model=MultinomialNB()
model.fit(X_train, y_train)
y_pred=model.predict(X_test)
```

```python
print("Accuracy", accuracy_score(y_test, y_pred))
print( classification_report(y_test, y_pred))
```
